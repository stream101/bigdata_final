### <p>(2) Use KD-tree to do the partition based on the weights </p>

### In each step, we find the medium point on either latitude or longitude to do the partition as KD-tree works. The process is done recursively until we reach the given level, for example 10 levels/1024 parts. children_df stores the partition information (partition id) for each station and children_region stores the border information for each partition. One partition id correspondes to one partition area/region.

```python
import pandas as pd

class KD_Partition(object):

    def __init__(self, df, levels):
        self.df = df
        self.levels = levels
        
        
    def get_id(self, bits):
        return sum(bit*(2**i) for bit, i in zip(bits, range(len(bits))))

    
    def execute(self, df, level, partition='latitude', N=90, S=-90, W=-180, E=180, bits=[]):
        if level <= 0:
            id = self.get_id(bits)
            df['id'] = id
            region = {'id': id, 'N': N, 'S': S, 'W': W, 'E': E, 'weight': df.weight.sum()}
            return df, region
        
        children_df = []
        children_region = []
        df = df.sort(partition)
        df['cumulative_weight'] = df.weight.cumsum()
        threshold = df.weight.sum()/2
        tree_id = 'tree' + str(level)
        df[tree_id] = df.cumulative_weight > threshold
        del df['cumulative_weight']
        half_df = df.groupby(tree_id)

        below_df = pd.DataFrame()
        above_df = pd.DataFrame()
        for flag, half in half_df:
            if flag: above_df = half
            else: below_df = half

        min_Value = above_df[partition].min()
        max_Value = below_df[partition].max()
        border = (max_Value + min_Value)/2

        for flag, child in half_df:
            if partition == 'latitude':
                if flag: S = border
                else: N = border
                child_df, child_region = self.execute(child, level-1, 'longtitude', N, S, W, E, [flag]+bits)
            else:
                if flag: W = border
                else: E = border
                child_df, child_region = self.execute(child, level-1, 'latitude', N, S, W, E, [flag]+bits)
        
            children_df.append(child_df)
            children_region.append(child_region)

        return pd.concat(children_df), children_region
```

```python
from mpl_toolkits.basemap import Basemap
import random

def color_map(lon, lat, Z, num, figsize=(15,10), alpha=1):
    lonmin = -180
    lonmax = 180
    latsmin = -80
    latsmax = 80;
    
    plt.figure(figsize=figsize,dpi=300)
    m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
        llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='c')
    m.drawcoastlines(color='lightgrey')
    m.drawmapboundary()    
    origin_color = Z
    map_color=random.sample(range(0, num), num)
    random_color=[0]*len(origin_color)
    for index in range(0, len(origin_color)-1):
        random_color[index]=map_color[origin_color[index]]
    x,y = m(np.array(lon), np.array(lat))
    color_map = m.scatter(x,y,c=np.array(random_color),marker='.', lw=0, alpha=alpha)
    m.colorbar(color_map)
```

### <p>(3) Merge the similar nodes based on the correlation coefficient</p>

```python
%%writefile calc_partition_vector.py
#!/usr/bin/python

import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr
import re,pickle,base64,zlib
import csv

class MRWeather(MRJob):
    
    def __init__(self, *args, **kwargs):
        super(MRWeather, self).__init__(*args, **kwargs)
        self.valid_stations = {}
    
    def configure_options(self):
        super(MRWeather,self).configure_options()
        self.add_file_option('--stationyear')
    
    def mapper_init(self):
        f= open(self.options.stationyear,'rb')
        reader = csv.reader(f)
        for line in reader:
            try:
                st = line[0]
                year = int(line[1])
                if st in self.valid_stations:
                    self.valid_stations[st].append(year)
                else:
                    self.valid_stations[st] = [year]
            except Exception as e: 
                pass
        f.close()
    
    def mapper(self, _, line):
        self.increment_counter('MrJob Counters','mapper-all',1)
        try: 
            elements=line.split(',')
            year = int(elements[2])/10000
            mean_temp = float(elements[3])
            if str(elements[0]) in self.valid_stations and elements[0] != '999999':
                if year in self.valid_stations[str(elements[0])] and year >= 1930 and year <= 2008 and mean_temp != 9999.9:
                    out = (int(elements[0]), (year, mean_temp))
                    yield out
        except Exception, e: 
            pass

    def reducer(self, station, info):
        
        self.increment_counter('MrJob Counters','reducer',1)
        info_list = list(info)
        mean_vector = [0] * 79
        mean_num = [0] * 79
        mean_sum = 0
        mean_cnt = 0
        num = len(info_list)
        for i in range(0, num):
            mean_vector[info_list[i][0]-1930] = mean_vector[info_list[i][0]-1930] + info_list[i][1]
            mean_num[info_list[i][0]-1930] = mean_num[info_list[i][0]-1930] + 1
            mean_sum = mean_sum + info_list[i][1]
            mean_cnt = mean_cnt + 1
        for i in range(0, 79):
            if mean_num[i] != 0:
                mean_vector[i] = mean_vector[i]/mean_num[i]
            else:
                mean_vector[i] = mean_sum / mean_cnt
        
        yield(station, mean_vector)   
        
if __name__ == '__main__':
    MRWeather.run()
```

```python
coff_info = [[0 for col in range(79)] for row in range(2048)]
coff_flag = [0]*2048
coff_id = range(0, 2048)
weight = [0]*2048
station_id = {}
station_weight = {}

data_in = open('weight_1024', 'r')
for line in data_in.readlines():
    words = line.split(',')
    words[1] = words[1].replace('\n', '')
    weight[int(words[1])] = weight[int(words[1])] + int(words[0])
data_in.close()

data_in = open('station_id_1024', 'r')
for line in data_in.readlines():
    words = line.split(',')
    words[1] = words[1].replace('\n', '')
    station_id[int(words[0])] = int(words[1])
data_in.close()


data_in = open('station_weight_1024', 'r')
for line in data_in.readlines():
    words = line.split(',')
    words[1] = words[1].replace('\n', '')
    station_weight[int(words[0])] = int(words[1])
data_in.close()

error_cnt = 0
data_in=open('partition_vector', 'r')
for line in data_in.readlines():
    try:
        word = line.split('\t')
        station_no = int(word[0].replace('"',''))
        word[1] = word[1].replace('[', '')
        word[1] = word[1].replace(']', '')
        word[1] = word[1].replace('\n', '')
        partition_vector = word[1].split(', ')
        for index in range(0, len(partition_vector)):
            coff_info[station_id[station_no]][index] = coff_info[station_id[station_no]][index] + float(partition_vector[index])*station_weight[station_no] 
    except Exception as e: 
        error_cnt = error_cnt + 1
        #print error_cnt
        
data_in.close()

for index2 in range(0, 1024):
    for index in range(0, 79):
        coff_info[index2][index] = coff_info[index2][index] / weight[index2]
    
```

```python
def update(x, y):
    coff_id[x]=y
    left = 2*x-2047
    right = 2*x-2048
    if right >= 0:
        coff_id[left] = y
        coff_id[right] = y
        update(left,y)
        update(right,y)
    return 1

def merge_partition(coff_threshold):
    
    for index in range(0, 1024):
        coff_flag[index] = 1
    for index in range(1024, 2048):
        coff_flag[index] = 0
        coff_info[index] = [0]*79

    coff_id = range(0, 2048)
        
    for index in range(1024, 2048):
        child_left = index*2-2047
        child_right = index*2-2048
        if coff_flag[child_left]==1 and coff_flag[child_right]==1:
            coff_value = np.corrcoef(coff_info[child_left], coff_info[child_right])
            if coff_value[1][0] >= coff_threshold:
                coff_flag[index] = 1
                for index2 in range(0, 79):
                    coff_info[index][index2] = (coff_info[child_left][index2]*weight[child_left] + coff_info[child_right][index2]*weight[child_right]) / (weight[child_left] + weight[child_right])
                weight[index] = weight[child_left] + weight[child_right]
                update(index, index)
    return 2048 - sum(coff_flag)
```
